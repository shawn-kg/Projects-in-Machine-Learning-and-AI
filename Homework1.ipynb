#Task 4: Optimizers

```python
model = tf.keras.Sequential([layers.Dense(303),layers.Dense(1)])
model.compile(loss = tf.losses.MeanSquaredLogarithmicError(),
                      optimizer = tf.optimizers.Adam())
model.fit(heart_features, heart_labels, epochs=10)
```

```python
model = tf.keras.Sequential([layers.Dense(303),layers.Dense(1)])
model.compile(loss = tf.losses.MeanSquaredLogarithmicError(),
                      optimizer = tf.optimizers.RMSprop())
model.fit(heart_features, heart_labels, epochs=10)
```
